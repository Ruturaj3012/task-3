```python
import requests
import folium
import webbrowser
import os

def get_geolocation():
    try:
        # Fetching data for the current user's IP address
        # Note: Some APIs require an access token for high-volume requests
        response = requests.get("https://ipinfo.io/json")
        data = response.json()
        
        # Extracting coordinates and details
        # IPinfo returns "loc" as "latitude,longitude"
        location = data.get("loc", "0,0").split(",")
        lat, lon = float(location[0]), float(location[1])
        city = data.get("city", "Unknown")
        region = data.get("region", "Unknown")
        ip = data.get("ip", "Unknown")
        
        print(f"IP detected: {ip}")
        print(f"Location: {city}, {region} ({lat}, {lon})")
        
        return lat, lon, f"IP: {ip}<br>Location: {city}, {region}"
    except Exception as e:
        print(f"Error fetching geolocation: {e}")
        return None

def create_map(lat, lon, popup_text):
    # Initialize the map centered at the coordinates
    geo_map = folium.Map(location=[lat, lon], zoom_start=12)
    
    # Add a marker for the location
    folium.Marker(
        [lat, lon],
        popup=popup_text,
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(geo_map)
    
    # Save the map as an HTML file
    map_filename = "geolocation_map.html"
    geo_map.save(map_filename)
    
    # Open the map in the default web browser
    full_path = os.path.abspath(map_filename)
    webbrowser.open(f"file://{full_path}")
    print(f"Map saved and opened: {full_path}")

if __name__ == "__main__":
    coords = get_geolocation()
    if coords:
        create_map(*coords)
```
